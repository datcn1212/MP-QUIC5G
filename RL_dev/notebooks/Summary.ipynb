```python
from itertools import cycle
import pandas

from matplotlib import pyplot as plt
import seaborn as sns; sns.set()
#loss/00
#rtt/05/output
#dynamic/low/output
#final/512kb/output
rtt = pandas.read_csv("/App/result/dynamic/high/output/rtt.csv")
ecf = pandas.read_csv("/App/result/dynamic/high/output/ecf.csv")
blest = pandas.read_csv("/App/result/dynamic/high/output/blest.csv")
peekaboo = pandas.read_csv("/App/result/dynamic/high/output/peekaboo.csv")
qlearning = pandas.read_csv("/App/result/dynamic/high/output/qlearning.csv")

for i in rtt.columns:
    rtt[i] = rtt[i].apply(lambda x: x*1.0/1000.0)
for i in ecf.columns:
    ecf[i] = ecf[i].apply(lambda x: x*1.0/1000.0)
for i in blest.columns:
    blest[i] = blest[i].apply(lambda x: x*1.0/1000.0)
for i in peekaboo.columns:
    peekaboo[i] = peekaboo[i].apply(lambda x: x*1.0/1000.0)
for i in qlearning.columns:
    qlearning[i] = qlearning[i].apply(lambda x: x*1.0/1000.0)

result = pandas.concat([(rtt), (ecf), (blest), (peekaboo), (qlearning)], axis=1)
import matplotlib.pyplot as plt
from IPython.core.pylabtools import figsize
figsize(9, 14)

sns.set_style("ticks", {'grid.linestyle': '--', 'font.family': ['sans-serif']})
#sns.set_style("ticks")
ax = sns.boxplot(data=result, showfliers=False, linewidth=2.0) #medianprops={'color': 'None'}
#sns.despine(left=True)
ax.tick_params(labelsize=25)
hatches = cycle(['/', 'x', '-', '.', '\\'])
colores = cycle(['blue', 'red', 'purple', 'orange', 'green'])
#result.plot.box(showfliers=False)
for i, patch in enumerate(ax.artists):
    # Boxes from left to right
    hatch = next(hatches)
    color = next(colores)
    patch.set_hatch(hatch)
    patch.set_edgecolor(color)
    patch.set_facecolor("white")
    
plt.ylabel("Download Time (s)", fontsize=25)
plt.ylim(top=50, bottom=10)
#plt.xlabel("High", fontsize=25)
#plt.title("0.0%")
#plt.xticks([])
#plt.yticks([])
#plt.show()
print(result.median())
```
